## subplots
subplot의 수를 미리 지정해놓는다.

```
# 데이터
data = np.arange(1, 51)

# 밑 그림
fig, axes = plt.subplots(2, 3) # subplot에 s를 붙여 만들고 미리 plot의 수를 지정해놓는다.

# axes에 간섭하여 그래프를 만든다.
axes[0, 0].plot(data)
axes[0, 1].plot(data * data)
axes[0, 2].plot(data ** 2)
axes[1, 0].plot( -data )
axes[1, 1].plot(data % 3)
axes[1, 2].plot(data ^ 3)

plt.tight_layout() # 보기 좋게 좀 더 타이트하게 만들어 주는 것.
plt.show()
```

## scatterplot (산점도)
점을 이용한 산점도 ( 어느 곳에 집중적으로 존재하는지 )

### 1-1 x, y color, area를 설정하여 준다.

```
x = np.random.rand(50)
y = np.random.rand(50)
# 색상
colors =  np.arange(50)
# 원의 면적
area = x * y * 1000
```

```
# s == size, c == color를 의미한다.
plt.scatter(x, y, s=area, c=colors)
plt.show()
```

### cmap and alpha
* cmap 그래프의 기본이 되는 색상을 조정
* 컬러 값을 모두 같게 가져갈 수 있다.
* alpha는 투명도를 나타낸다. (0~1로 지정)

```
# figure를 사용해 새로 만들어준다.
plt.figure(figsize=(12, 6)) # 사이즈를 조정해준다.

plt.subplot(1, 3, 1)
plt.scatter(x, y, s=area, cmap='blue', alpha=0.1)
plt.title('alpha=0.1')

plt.subplot(1, 3, 2)
plt.scatter(x, y, s=area, cmap='blue', alpha=0.5)
plt.title('alpha=0.5')

plt.subplot(1, 3, 3)
plt.scatter(x, y, s=area, cmap='blue', alpha=1.0)
plt.title('alpha=1.0')

plt.show()
```

### 가로로 되어 있는 막대 그래프 (barhplot)
xtick에 넣을 단어가 길어 글자가 곂칠 때, 보기 좋게 하기 위해 x,y tick을 바꾼 것

```
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

plt.figure(figsize = (6, 3))
plt.barh(x, y, align='center', alpha=0.6, color='red')

plt.title('Subject')

plt.yticks(x) # barplot에서 xticks로 했다면 barhplot에서는 ytick를 사용한다.

plt.ylabel('Number of Students')

plt.show()
```

```
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
x = np.arange(len(x_label))
y_1 = [66, 80, 60, 50, 80, 10]
y_2 = [55, 90, 40, 60, 70, 20]

# subplots 생성
fig, axes = plt.subplots()
# 넓이 지정
width = 0.35
# 넓이 설정 ( x, y, 너비 )
axes.bar(x - width/2, y_1, width, align='center', alpha = 0.5 ) # x-width/2 만큼 막대를 이동시켜 2개의 막대가 겹치지 않고 비교하며 볼 수 있도록 해줌.
axes.bar(x + width/2, y_2, width, align='center', alpha = 0.8 )
# xtick 설정
plt.xticks(x)
axes.set_xticklabels(x_label)
plt.ylabel('Number of Students')
plt.title('Subjects')

plt.legend(['john', 'peter'])

plt.show()
```

## Lineplot (선 그래프)

### 기본적인 선 그래프 그리기

```
x = np.arange(0, 10, 0.1)
y = 1 + np.sin(x)

plt.plot(x, y) # 핵심

plt.xlabel('x value', fontsize=15)
plt.ylabel('y value', fontsize=15)

plt.title('sin graph', fontsize=30)

plt.grid()
plt.show()
```

### 두 개의 선 그래프

```
x = np.arange(0, 10, 0.1)
y1 = 1 + np.sin(x)
y2 = 1 + np.cos(x)

plt.plot(x, y1, label='1+sin', color='blue', alpha=0.3)
plt.plot(x, y2, label='1+cos', color='red', alpha=0.7)

plt.xlabel('x value', fontsize=15)
plt.ylabel('x value', fontsize=15)

plt.title('sin and cos graph', fontsize=18)

plt.grid()
plt.legend()
plt.show()
```

### 선 그래프 라인, 마커 변경하기

```
# 선 그래프의 라인 변경 linestyle
plt.figure(figsize=(12,9))
x = np.arange(0, 10, 0.1)
y1 = 1 + np.sin(x)
y2 = 1 + np.cos(x)

plt.plot(x, y1, label='1+sin', color='blue', alpha=1, linestyle=':')
plt.plot(x, y2, label='1+cos', color='red', alpha=0.7, linestyle='-.')

plt.xlabel('x value', fontsize=15)
plt.ylabel('x value', fontsize=15)

plt.title('sin and cos graph', fontsize=18)

plt.grid()
plt.legend()

plt.show()
```

```
# 선 그래프의 마커 변경
x = np.arange(0, 10, 0.1)
y1 = 1 + np.sin(x)
y2 = 1 + np.cos(x)

plt.plot(x, y1, label='1+sin', color='blue', alpha=1, marker='+')
plt.plot(x, y2, label='1+cos', color='red', alpha=0.7, marker='v')

plt.xlabel('x value', fontsize=15)
plt.ylabel('x value', fontsize=15)

plt.title('sin and cos graph', fontsize=18)

plt.grid()
plt.legend()
plt.show()
```

## AREA plot (filled area)

## Pie chart
점유율을 표시할 때 주로 사용한다.
(4~5개의 적은 수는 괜찮지만 그 이상일 때는 barchart가 낫다.)
* explode: 툭 튀어나온 비율 ( 피자 한 조각을 때어내는 표현 )
* autopct: 퍼센트율 자동 표기
* shadow: 그림자 표기
* startangle: 파이를 그리기 시작하는 각도

texts와 autotexts의 리턴 인자를 받는다.
* text: label에 대한 텍스트 효과
* autotexts: 파이 위에 그려지는 텍스트 효과

```
labels = ['피카츄', '꼬부기', '파이리', '이상해씨']
sizes = [30, 30, 20, 10] # 비율이기 때문에 꼭 100을 맞출 필요는 없다.
explode = (0, 0, 0.2, 0)

# texts, autotexts를 사용해 텍스트 스타일링 적용
patches, texts, autotexts = plt.pie(sizes,
                                    explode=explode, # explode를 사용하기 위해서는 넣어줘야한다.
                                    labels=labels,
                                    autopct='%1.1f%%', # %1.1f로 소수점 첫 번째 자리까지 나오도록 표기하며 %%를 사용해 %를 넣어준다.
                                    shadow=True,
                                    startangle=45)

# texts를 사용해 label 스타일 적용
for i in texts:
    i.set_fontsize(13)
    i.set_color('c')
# autotexts를 사용해 pie의 스타일 적용
for i in autotexts:
    i.set_fontsize(15)
    i.set_color('w')
plt.title('포컷몬 선호도 조사', fontsize=30)

plt.show()
```

### contour3D ( 등고선 )
높낮이를 이욯해

```
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
x, y= np.meshgrid(x, y) # x, y를 사용하여 2차원 그리드 좌표를 만들어준다. / grid란 좌표에 나타나는 선.

z = np.sin(np.sqrt(x**2 + y**2)) # sqrt: 루트

fig = plt.figure(figsize=(12, 6))
ax = plt.axes(projection='3d')

ax.contour3D(x, y, z, 20, cmap='Reds')

plt.title('등고선 그래프')
plt.show()
```

## heatmap ( 히트 맵 )
* 색으로 표현 가능한 데이터를 열분포(지역을 나눠 수치에 따라 다른 색으로 표현) 형태로 출력
사용
* pivot table을 돌려서 x축 y축을 기준으로 value를 **시각화** 할 때
* 데이터의 **상관관계**를 볼 때

```
data = np.random.rand(10, 12) # 12 개씩 10개의 arange데이터 생성

sns.heatmap(data)

plt.show()
```

```
# heatmap 위에 숫자로 수치가 나타낼 수 있음
data = np.random.rand(10, 12) # 12 개씩 10개의 arange데이터 생성

sns.heatmap(data, annot=True)

plt.show()
```

위의 그래프를 통해 임의의 x축과 y축이 만났을 때 어떤 value가 나타나는지 시각적으로 확인이 가능하다

### pivot table 을 사용하여 heatmap 그래프 그리기

```
tips = sns.load_dataset('tips')
pivot = tips.pivot_table(index='day', columns='size', values='tip')

pivot
```

```
sns.heatmap(pivot, annot=True, cmap='Blues') # cmap은 그래프의 기본적인 색상 구성을 조정할 수 있음

plt.ylabel('day', rotation=0)
plt.yticks(rotation=0)
plt.show()
```

### 상관관 ( correlation )
* corr() / 데이터의 상관관계를 표현하는 함수
* 양, 음의 상관관계를 통해 결과를 유추해낼 수 있다.

```
titanic = sns.load_dataset('titanic')
titanic
```

```
# column 들 간의 상관관계를 표현한다.
titanic.corr()
```

```
# 상관관계가 나타난 데이터를 사용해 heatmap 그래프를 그린다.
# surived와 male의 상관관계가 음의 상관관계가 -0.56으로 높은 수치인 것을 보고 male은 surived가 낮다는 것을 알 수 있다.
data = titanic.corr()

sns.heatmap(data, annot=True, cmap='BrBG')

plt.show()
```

## lmplot (LM)
* column간 선형관계 확인에 용이
* outlier 짐작 가능

```
# 선형관계의 선을 자동으로 그려줌
tips = sns.load_dataset('tips')

sns.lmplot(x='total_bill', y='tip', height=8, data=tips)
plt.show()
```

```
# hue를 통해 smoker를 기준으로 비교 / 비흡연자가 tip을 더 많이 주는 경향이 있음을 확인
sns.lmplot(x='total_bill', y='tip', hue='smoker', height=7, data=tips)
plt.show()
```

### 여러개의 그래프

```
# col을 사용해 day를 기준으로 여러개의 그래프를 그림
sns.lmplot(x='total_bill', y='tip', col='day', data=tips)
plt.show()
```

```
# col을 통해 생긴 그래프들을 보기 좋게 나열하기 위해 col_wrap 사용
sns.lmplot(x='total_bill', y='tip', col='day', height=3, col_wrap=2, data=tips)
plt.show()
```

```
# hue와 col을 동시 사용
sns.lmplot(x='total_bill', y='tip', hue='smoker', col='day', height=5, col_wrap=2, data=tips)
plt.show()
```

## relplot
* lmplot과 유사하나 선형관계는 그려주지 않는다.

```
tips = sns.load_dataset('tips')

sns.relplot(x='total_bill', y='tip', height=7, data=tips)
plt.grid()
plt.show()
```

```
sns.relplot(x='total_bill', y='tip', hue='day', height=7, data=tips)

plt.grid()
plt.show()
```

```
sns.relplot(x='total_bill', y='tip', hue='day', col='time', height=7, data=tips)

plt.show()
```

```
# col 뿐만 아니라 row 옵션을 넣을 수도 있다.
sns.relplot(x='total_bill', y='tip', hue='day', row='sex', col='time', data=tips)
plt.show()
```

## joint plot
* scatter(산점도)와 histogram(분포)를 동시에 그림
* 숫자형 데이터만 사용 가능

```
tips = sns.load_dataset('tips')

sns.jointplot(x='total_bill', y='tip', height=7, data=tips)
plt.show()
```

### 그래프에 다양한 옵션을 넣을 수가 있다.
* kind='scatter'
* 선형관계(regression line) 추가 : kind='reg'
* kind='resid'
* 등고선 모양의 그래프 : kind='kde'
* hex모양으로 이뤄진 그래프 : kind='hex'

```
# 선형관계 그리기
sns.jointplot(x='total_bill', y='tip', height=7, kind='reg', data=tips)
plt.show()
```

```
# hex 밀도 확인
sns.jointplot(x='total_bill', y='tip', height=7, kind='hex', data=tips)
plt.show()
```

```
# 등고선 그리기
sns.jointplot(x='total_bill', y='tip', height=7, kind='kde', data=tips)
plt.show()
```
